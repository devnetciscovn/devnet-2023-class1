READ MESSAGES

```python
import requests
from Webex_credentials import base_url, token, roomId
```

```python
url = base_url + '/v1/messages'

webex_params = {
    'roomId': roomId,
    'sortBy': 'lastactivity'
}

webex_headers = {
    'Authorization': f'Bearer {token}'
}

#print(url)
#print(webex_headers)
#print(webex_params)
```

```python
response = requests.get(url, headers=webex_headers, params=webex_params)
print(response.status_code)
```

```python
response.json()["items"]
```

```python
for item in response.json()["items"]:
    print(item)
```

```python
for item in reversed(response.json()["items"]):
    print(item["personEmail"], item["text"], sep="\n\t")
```

CREATE A MESSAGE

```python
import requests
from Webex_credentials import base_url, token, roomId
```

```python
url = f'{base_url}/v1/messages'
webex_headers = {
    'Authorization': f'Bearer {token}'
}

payload = {
    #"toPersonEmail": "anhtuyetpt@gmail.com",
    'roomId': roomId,
    "text": "Hello from Python"
}
print(url)
print(webex_headers)
print(payload)
```

```python
response = requests.post(url, headers=webex_headers, json=payload)
print(response.status_code)
```

```python
import json

output = json.dumps(response.json(),indent=8)

print(output)
```
